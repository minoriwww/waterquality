```python
from PIL import Image
from matplotlib import cm
from numpy import average, dot, linalg
 
# 对图片进行统一化处理
def get_thum(image, size=(64,64), greyscale=False):
    # 利用image对图像大小重新设置, Image.ANTIALIAS为高质量的
    image = Image.fromarray((image * 255).astype(np.uint8))#Convert opencv image format to PIL image format
    image = image.resize(size, Image.ANTIALIAS)
#     image = cv2.resize(image,None,fx=0.5,fy=0.5)
    if greyscale:
        # 将图片转换为L模式，其为灰度图，其每个像素用8个bit表示
        image = image.convert('L')
    return image
 
# 计算图片的余弦距离
def image_similarity_vectors_via_numpy(image1, image2):
    image1 = get_thum(image1)
    image2 = get_thum(image2)
    
    images = [image1, image2]
    vectors = []
    norms = []
    for image in images:
        vector = []
        for pixel_tuple in image.getdata():
            vector.append(average(pixel_tuple))
        vectors.append(vector)
        # linalg=linear（线性）+algebra（代数），norm则表示范数
        # 求图片的范数？？
        norms.append(linalg.norm(vector, 2))
    a, b = vectors
    a_norm, b_norm = norms
#     print( a, b, a_norm, b_norm)
    # dot返回的是点积，对二维数组（矩阵）进行计算
    res = dot(a / a_norm, b / b_norm)

def split_image(img, part_size = (128, 128)):
    w, h = img.size
    pw, ph = part_size
    assert w % pw == h % ph == 0
    return [img.crop((i, j, i+pw, j+ph)).copy() for i in range(0, w, pw) for j in range(0, h, ph)]

def hist_similar(lh, rh):
    assert len(lh) == len(rh)
    hist = sum(1 - (0 if l == r else float(abs(l-r))/max(l,r))for l, r in zip(lh, rh))/len(lh)
    return hist
 
# 计算相似度
def calc_similar(li, ri):
    li = Image.fromarray((li * 255).astype(np.uint8))
    ri = Image.fromarray((ri * 255).astype(np.uint8))
#     calc_sim = hist_similar(li.histogram(), ri.histogram())
#     return calc_sim
    return sum(hist_similar(l.histogram(), r.histogram()) for l, r in zip(split_image(li), split_image(ri))) / 16.0

def calc_similar_whole_image(li, ri):
    li = Image.fromarray((li * 255).astype(np.uint8))
    ri = Image.fromarray((ri * 255).astype(np.uint8))
#     calc_sim = hist_similar(li.histogram(), ri.histogram())
#     return calc_sim
    return hist_similar(li.histogram(), ri.histogram())
   

# 差异值哈希算法
def dhash(image, hashSize=8):
    # resize the input image, adding a single column (width) so we
	# can compute the horizontal gradient
	resized = cv2.resize(image, (hashSize + 1, hashSize))
 
	# compute the (relative) horizontal gradient between adjacent
	# column pixels
	diff = resized[:, 1:] > resized[:, :-1]
 
	# convert the difference image to a hash
	return str(sum([2 ** i for (i, v) in enumerate(diff.flatten()) if v]))


# 计算两个哈希值之间的差异
def campHash(hash1, hash2):
    n = 0
    # hash长度不同返回-1,此时不能比较
    if len(hash1) != len(hash2):
        return -1
    # 如果hash长度相同遍历长度
    for i in range(len(hash1)):
        if hash1[i] != hash2[i]:
            n = n+1
    return n
```

```python
import itertools 
def hist_similar_random_all_x(X_train, X_test, nb=3):
    X = np.concatenate((X_train, X_test), axis=0)
    selected_arr = np.random.randint(0,X.shape[0],nb)
    res = []
    for i,j in itertools.combinations(selected_arr,2):
        res.append(calc_similar(X[i], X[j]))
    return res

hist_arr = hist_similar_random_all_x(X_train, X_test, 30)

sns.distplot(hist_arr);
```

```python
import joblib
def hist_dist_random_all_x_y(X_train, X_test, y_train, y_test, nb=3):
                     
    X = np.concatenate((X_train, X_test), axis=0)
    y = np.concatenate((y_train, y_test), axis=0)
    scaler_val = joblib.load('MaxAbsScaler.pkl')
    y_test_inverse = scaler_val.inverse_transform(y.reshape(-1, 1))
    
    selected_arr = np.random.randint(0,X.shape[0],nb)
    res_x = []
    res_y = []
    for i,j in itertools.combinations(selected_arr,2):
        res_x.append(calc_similar(X[i], X[j]))
        res_y.append(abs(y_test_inverse[i][0]-y_test_inverse[j][0]))
    return res_x, res_y

res_x, res_y = hist_dist_random_all_x_y(X_train, X_test, y_train, y_test, 100)

import scipy 
print(scipy.stats.pearsonr(res_x, res_y))
# the low p-value and negative pearson correlation coefficient indicates that the histogram similarity and label similarity have some linear relationship.
print(scipy.stats.spearmanr(res_x, res_y))

plt.scatter(res_x, res_y)
plt.xlabel("Pairwise histogram similarity, Xi and Xj; 2 same images' similarity is 1.0")
plt.ylabel("Pairwise similarity of label, abs(yi - yj)")
plt.show()
plt.savefig('similarity.png')
```

```python

def hist_RGB_per_image(image = X_train[2]):
    image = Image.fromarray((image * 255).astype(np.uint8))
    histogram = image.histogram()

    def getRed(redVal):
        return '#%02x%02x%02x' % (redVal, 0, 0)
    def getGreen(greenVal):
        return '#%02x%02x%02x' % (0, greenVal, 0)
    def getBlue(blueVal):
        return '#%02x%02x%02x' % (0, 0, blueVal)
    # Take only the Red counts
    l1 = histogram[0:256]
    # Take only the Blue counts
    l2 = histogram[256:512]
    # Take only the Green counts
    l3 = histogram[512:768]
    # plt.figure(0)
    # # R histogram
    # for i in range(0, 256):
    #     plt.bar(i, l1[i], color = getRed(i), edgecolor=getRed(i), alpha=0.3)
    # # G histogram
    # plt.figure(1)
    # for i in range(0, 256):
    #     plt.bar(i, l2[i], color = getGreen(i), edgecolor=getGreen(i),alpha=0.3)
    # # B histogram
    # plt.figure(2)
    # for i in range(0, 256):
    #     plt.bar(i, l3[i], color = getBlue(i), edgecolor=getBlue(i),alpha=0.3)
    # plt.show()
    return l1, l2, l3

def hist_RGB_stat(X_train, X_test, y_train, y_test, nb=30):
    X = np.concatenate((X_train, X_test), axis=0)
    y = np.concatenate((y_train, y_test), axis=0)
    scaler_val = joblib.load('MaxAbsScaler.pkl')
    y_test_inverse = scaler_val.inverse_transform(y.reshape(-1, 1))
    
    selected_arr = np.random.randint(0,X.shape[0],nb)
    argmax_arr_R, argmax_arr_G, argmax_arr_B = [], [], []
    count_nonzero_arr_R, count_nonzero_arr_G, count_nonzero_arr_B = [], [], []
    selected_y = []
    for i in selected_arr:
        l1, l2, l3 = hist_RGB_per_image(X[i])
        argmax_arr_R.append(np.argmax(l1))
        argmax_arr_G.append(np.argmax(l2))
        argmax_arr_B.append(np.argmax(l3))
        count_nonzero_arr_R.append(np.count_nonzero(l1))
        count_nonzero_arr_G.append(np.count_nonzero(l2))
        count_nonzero_arr_B.append(np.count_nonzero(l3))
        selected_y.append(y_test_inverse[i][0])
    return argmax_arr_R, argmax_arr_G, argmax_arr_B, count_nonzero_arr_R, count_nonzero_arr_G, count_nonzero_arr_B, selected_y
    
# print(np.argmax(l1), np.argmax(l2), np.argmax(l3))
# print(np.count_nonzero(l1), np.count_nonzero(l2), np.count_nonzero(l3))
argmax_arr_R, argmax_arr_G, argmax_arr_B, count_nonzero_arr_R, count_nonzero_arr_G, count_nonzero_arr_B,selected_y = hist_RGB_stat(X_train, X_test, y_train, y_test, 60)
plt.figure(0)
# plt.plot(argmax_arr_R, color = 'r');
sns.kdeplot(argmax_arr_R, color = 'r')
# plt.plot(argmax_arr_G, color = 'g');
# plt.plot(argmax_arr_B, color = 'b');
sns.kdeplot(argmax_arr_G, color = 'g');
sns.kdeplot(argmax_arr_B, color = 'b');
plt.show()
plt.close()

plt.figure(1)
sns.kdeplot(count_nonzero_arr_R, color = 'r');
sns.kdeplot(count_nonzero_arr_G, color = 'g');
sns.kdeplot(count_nonzero_arr_B, color = 'b');
plt.show()
```
